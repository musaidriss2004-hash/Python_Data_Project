# Basics: Exercise Skill Investigation

Coding

*   Loop through the job_roles
*   Verify if all my_skills are inside each job_roles['skills']
*   If True (or qualified), add to qualified list of jobs





```python
my_skills = ['Python', 'SQL', 'Excel']
```

```python
job_roles = [
    {'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']},
    {'role': 'Data Scientist', 'skills': ['Python', 'R', 'Machine Learning', 'Deep Learning']},
    {'role': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn']},
    {'role': 'Data Engineer', 'skills': ['Python', 'Apache Spark', 'Hadoop', 'SQL']},
    {'role': 'Business Intelligence Analyst', 'skills': ['Python', 'SQL', 'Tableau', 'Power BI', 'Excel']},
    {'role': 'Quantitative Analyst', 'skills': ['R', 'Python', 'MATLAB', 'Statistics']},
    {'role': 'Operations Analyst', 'skills': ['Python', 'SQL', 'Data Visualization', 'Process Improvement']},
    {'role': 'Database Administrator', 'skills': ['SQL', 'Oracle', 'MySQL', 'Database Management']},
    {'role': 'AI Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Computer Vision']},
    {'role': 'Statistician', 'skills': ['R', 'SAS', 'Python', 'Statistical Modeling']} ]
```

Below are 3 different versions to get answer

my_list.append(item)
Means: Take item and put it at the end of my_list.

```python
qualified_roles = []

for job in job_roles:
    qualified = True

    for skill in my_skills:
        if skill not in job['skills']:
            qualified = False
            break

    if qualified:
        qualified_roles.append(job['role'])

print(qualified_roles)
```

```python
qualified_roles = []

for job in job_roles:
  if all(skill in job['skills'] for skill in my_skills):
    qualified_roles.append(job['role'])

print(qualified_roles)
```

```python
qualified_roles = [job['role'] for job in job_roles if all(skill in job['skills'] for skill in my_skills)]
print(qualified_roles)
```

Basics Classes Chapter 21 in Notes

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol= "£"):
    self.base_salary = base_salary
    self.bonus_rate = 0.1
    self.symbol = symbol
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus = (self.total_salary - self. base_salary)

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary = BaseSalary(100000)
salary.base_salary
```

```python
salary
```

```python
value = 100000
f'{value:,.2f}' # ,.2f means 2 decimal places and with comma
```

```python
salary.show_salary()
```

```python
salary.show_bonus()
```
